Count Method:

The Python standard version of count() in a list will only search for an exact match of the word. Therefore we would have to lowercase all items/words in the list to get an accurate count. The NLTK version of the method count() on a text acts the same way but takes punctuation into account! This is because the text itself has been tokenized with punctuation being seperated by the word before it.

```python
#NLTK
print("NLTK \n")
python_counter = text1.count('sea')
print(python_counter)

python_counter = text1.count('sea.')
print(python_counter)

python_counter = text1.count('seas')
print(python_counter)

python_counter = text1.count('Seas')
print(python_counter)

python_counter = text1.count('Sea')
print(python_counter)

python_counter = text1.count('SEA')
print(python_counter)

#PYTHON
print("\nPYTHON\n")
tester = ["sea.", "air", "land", "sea", "sea", "seas", "SEAS"]
x = tester.count("sea")
print(x)
x = tester.count("Sea")
print(x)
```

```python
from nltk.corpus import (
    genesis,
    gutenberg,
    inaugural,
    nps_chat,
    treebank,
    webtext,
    wordnet,
)
from nltk.probability import FreqDist
from nltk.text import Text
from nltk.util import bigrams


from urllib import request
url = "http://www.gutenberg.org/files/2554/2554-0.txt"
response = request.urlopen(url)
raw = response.read().decode('utf8')

tokens = word_tokenize(raw)
print(tokens[:10])

print('\n')

sentences = sent_tokenize(raw)
#print(sentences[:5])

blocker = 0
for sentence in sentences:
    print(sentence)
    blocker += 1
    if blocker > 4:
      break
```

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()

singles = [stemmer.stem(token) for token in tokens]
print(' '.join(singles))
```

Stem:punish   Lemma:punishment

```python
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()

singles = [wnl.lemmatize(token) for token in tokens]
print(' '.join(singles))

```
