```python
from transformers import BertForQuestionAnswering, AutoTokenizer, Trainer, TrainingArguments
from datasets import Dataset

# Load the pre-trained BERT model and ensure the tokenizer is a fast tokenizer
model_name = 'bert-base-uncased'
tokenizer = AutoTokenizer.from_pretrained(model_name, use_fast=True)
model = BertForQuestionAnswering.from_pretrained(model_name)

# Define the data
data_list = [
    {
        "question": "Did Neymar play for Barcelona?",
        "context": "Neymar made his professional debut with Santos in 2009, and in 2011, he helped them win their first Copa Libertadores title since 1963. Neymar transferred to Barcelona in 2013, and won two La Liga titles, three Copa del Rey, and the UEFA Champions League in 2015.",
        "answers": {
            "text": ["Neymar transferred to Barcelona in 2013"],
            "answer_start": [23]
        }
    },
    {
        "question": "Where is Neymar from?",
        "context": "Debuting for Brazil aged 18, Neymar is the nation's all-time top goalscorer, with 79 goals in 128 matches.",
        "answers": {
            "text": ["Brazil"],
            "answer_start": [1]
        }
    },
    {
        "question": "Has Neymar won a world cup?",
        "context": "Helping Brazil to a runner-up finish at the 2021 Copa América, he was jointly awarded Best Player. In the 2022 World Cup, he became the third Brazilian player to score in three World Cups, after Pelé and Ronaldo. Neymar has won a record six Samba Gold awards.",
        "answers": {
            "text": ["No answer"],
            "answer_start": [-1]
        }
    },
    {
        "question": "Has Neymar won the Copa Libertadores?",
        "context": "Neymar made his professional debut with Santos in 2009, and in 2011, he helped them win their first Copa Libertadores in nearly 50 years.",
        "answers": {
            "text": ["Neymar has won the Copa Libertadores"],
            "answer_start": [12]
        }
    },
    {
        "question": "When did Neymar join PSG?",
        "context": "Neymar joined Paris Saint-Germain (PSG) in 2017 in a transfer costing â‚¬222 million, making him the most expensive player ever.",
        "answers": {
            "text": ["Neymar joined Paris Saint-Germain (PSG) in 2017"],
            "answer_start": [0]
        }
    },
    {
        "question": "Has Neymar won the league in France?",
        "context": "Neymar joined Paris Saint-Germain (PSG) in 2017 in a transfer costing 222 million, making him the most expensive player ever. There, he won Ligue 1 Player of the Year, won five Ligue 1 titles, and was integral to PSG being runners-up in the 2019-20 Champions League.",
        "answers": {
            "text": ["Neymar won in France with PSG"],
            "answer_start": [20]
        }
    },
    {
        "question": "Does Neymar get injured frequently?",
        "context": "He also ranks as PSG's fourth-highest all-time top goalscorer, despite reoccurring injuries consistently disrupting his playing time.",
        "answers": {
            "text": ["Yes, he has re-occurring injuries consistently disrupting his playing time"],
            "answer_start": [9]
        }
    },
    {
        "question": "Is Neymar influential?",
        "context": "In 2017, Time included him in its annual list of the 100 most influential people in the world.",
        "answers": {
            "text": ["Yes he has been featured in Time Magazines top 100 more influential people!"],
            "answer_start": [0]
        }
    },
    {
        "question": "What club did Messi leave in 2021?",
        "context": "In August 2021, Lionel Messi left FC Barcelona after over two decades at the club, due to financial and structural obstacles.",
        "answers": {
            "text": ["FC Barcelona"],
            "answer_start": [3]
        }
    },
    {
        "question": "Where did Cristiano Ronaldo move in 2021?",
        "context": "Cristiano Ronaldo rejoined Manchester United in 2021 after leaving Juventus, marking a significant return to the English club where he had previously made his name.",
        "answers": {
            "text": ["Manchester United"],
            "answer_start": [0]
        }
    },
    {
        "question": "Who won the Ballon d'Or in 2018?",
        "context": "The 2018 Ballon d'Or was awarded to Luka Modric, breaking the decade-long dominance of Lionel Messi and Cristiano Ronaldo in the award.",
        "answers": {
            "text": ["Luka Modric"],
            "answer_start": [2]
        }
    },
    {
        "question": "What team did Zlatan Ibrahimovic play for in MLS?",
        "context": "Zlatan Ibrahimovic played for LA Galaxy in the MLS, where he scored 52 goals in 56 appearances.",
        "answers": {
            "text": ["LA Galaxy"],
            "answer_start": [0]
        }
    },
    {
        "question": "Who is the highest-paid athlete in 2022?",
        "context": "As of 2022, the highest-paid athlete is Conor McGregor, primarily due to his earnings from the UFC and various endorsement deals.",
        "answers": {
            "text": ["Conor McGregor"],
            "answer_start": [4]
        }
    },
    {
        "question": "What year did the Chicago Bulls win their last NBA championship?",
        "context": "The Chicago Bulls last won the NBA championship in 1998, which marked their sixth title in eight years.",
        "answers": {
            "text": ["1998"],
            "answer_start": [0]
        }
    },
    {
        "question": "Which country hosted the 2016 Summer Olympics?",
        "context": "The 2016 Summer Olympics were held in Rio de Janeiro, Brazil, marking the first time the Olympics were held in South America.",
        "answers": {
            "text": ["Brazil"],
            "answer_start": [2]
        }
    },
    {
        "question": "What record did Usain Bolt set at the 2008 Olympics?",
        "context": "At the 2008 Beijing Olympics, Usain Bolt set a world record in the 100 meters with a time of 9.69 seconds.",
        "answers": {
            "text": ["world record in the 100 meters"],
            "answer_start": [4]
        }
    },
    {
        "question": "Which team won the 2020 UEFA Champions League?",
        "context": "Bayern Munich clinched the 2020 UEFA Champions League title, defeating Paris Saint-Germain in the final.",
        "answers": {
            "text": ["Bayern Munich"],
            "answer_start": [0]
        }
    },
    {
        "question": "What film won the Best Picture Oscar in 2019?",
        "context": "The film 'Parasite' won the Best Picture at the 2019 Academy Awards, becoming the first non-English language film to win the award.",
        "answers": {
            "text": ["Parasite"],
            "answer_start": [1]
        }
    },
    {
        "question": "Who is the all-time top goalscorer for Portugal?",
        "context": "Cristiano Ronaldo became the all-time top goalscorer for Portugal during the 2018 FIFA World Cup.",
        "answers": {
            "text": ["Cristiano Ronaldo"],
            "answer_start": [0]
        }
    },
    {
        "question": "When did Serena Williams win her first Wimbledon title?",
        "context": "Serena Williams won her first Wimbledon title in 2002, marking the beginning of her dominance in women's tennis.",
        "answers": {
            "text": ["2002"],
            "answer_start": [0]
        }
    },
    {
        "question": "What company did Elon Musk found in 2002?",
        "context": "Elon Musk founded SpaceX in 2002 with the goal of reducing space transportation costs to enable the colonization of Mars.",
        "answers": {
            "text": ["SpaceX"],
            "answer_start": [0]
        }
    },
    {
        "question": "Which car manufacturer won the Formula 1 constructors' championship in 2020?",
        "context": "Mercedes won the Formula 1 constructors' championship in 2020, continuing their dominance in the sport.",
        "answers": {
            "text": ["Mercedes"],
            "answer_start": [0]
        }
    },
    {
        "question": "Where is the headquarters of the United Nations?",
        "context": "The headquarters of the United Nations is located in New York City, established there in 1952.",
        "answers": {
            "text": ["New York City"],
            "answer_start": [8]
        }
    },
    {
        "question": "What novel did J.K. Rowling publish in 1997?",
        "context": "J.K. Rowling published 'Harry Potter and the Philosopher's Stone' in 1997, which became a global phenomenon.",
        "answers": {
            "text": ["Harry Potter and the Philosopher's Stone"],
            "answer_start": [3]
        }
    },
    {
        "question": "Who directed the movie 'Inception'?",
        "context": "The movie 'Inception', released in 2010, was directed by Christopher Nolan.",
        "answers": {
            "text": ["Christopher Nolan"],
            "answer_start": [8]
        }
    },
    {
        "question": "What is the capital of Japan?",
        "context": "Tokyo, the bustling capital of Japan, is known for its towering skyscrapers and vibrant culture.",
        "answers": {
            "text": ["Tokyo"],
            "answer_start": [0]
        }
    },
    {
        "question": "Who discovered penicillin?",
        "context": "Penicillin was discovered in 1928 by Alexander Fleming, marking a major advancement in antibacterial treatment.",
        "answers": {
            "text": ["Alexander Fleming"],
            "answer_start": [6]
        }
    },
    {
        "question": "Which animal is known as the ship of the desert?",
        "context": "The camel, often referred to as the ship of the desert, is vital for transport across arid environments.",
        "answers": {
            "text": ["The camel"],
            "answer_start": [0]
        }
    },
    {
        "question": "What is the tallest mountain in the world?",
        "context": "Mount Everest is recognized as the tallest mountain in the world, peaking at 8,848 meters above sea level.",
        "answers": {
            "text": ["Mount Everest"],
            "answer_start": [0]
        }
    },
    {
        "question": "When was the iPhone first introduced?",
        "context": "The first iPhone was introduced by Apple in 2007, revolutionizing the smartphone market.",
        "answers": {
            "text": ["2007"],
            "answer_start": [8]
        }
    },
    {
        "question": "Who wrote 'Pride and Prejudice'?",
        "context": "The novel 'Pride and Prejudice', a classic in English literature, was written by Jane Austen.",
        "answers": {
            "text": ["Jane Austen"],
            "answer_start": [13]
        }
    },
    {
        "question": "What is the chemical symbol for water?",
        "context": "Water is often referred to by its chemical symbol, H2O, consisting of two hydrogen atoms and one oxygen atom.",
        "answers": {
            "text": ["H2O"],
            "answer_start": [9]
        }
    },
    # {
    #     "question": "What year did the Titanic sink?",
    #     "context": "The Titanic tragically sank on its maiden voyage in 1912 after hitting an iceberg in the North Atlantic.",
    #     "answers": {
    #         "text": ["1912"],
    #         "answer_start": [58]
    #     }
    # },
    # {
    #     "question": "Who invented the light bulb?",
    #     "context": "Thomas Edison is often credited with the invention of the light bulb, although he improved upon previous designs rather than invented it.",
    #     "answers": {
    #         "text": ["Thomas Edison"],
    #         "answer_start": [0]
    #     }
    # },
    # {
    #     "question": "Which planet is known as the Red Planet?",
    #     "context": "Mars is commonly referred to as the Red Planet due to its reddish appearance caused by iron oxide on its surface.",
    #     "answers": {
    #         "text": ["Mars"],
    #         "answer_start": [0]
    #     }
    # },
    # {
    #     "question": "Who painted the Mona Lisa?",
    #     "context": "The Mona Lisa, one of the most famous paintings in the world, was painted by Leonardo da Vinci.",
    #     "answers": {
    #         "text": ["Leonardo da Vinci"],
    #         "answer_start": [63]
    #     }
    # },
    # {
    #     "question": "What is the hardest natural substance on Earth?",
    #     "context": "Diamond is considered the hardest natural substance on Earth, widely used in industrial applications.",
    #     "answers": {
    #         "text": ["Diamond"],
    #         "answer_start": [0]
    #     }
    # },
    # {
    #     "question": "What year was the United Nations founded?",
    #     "context": "The United Nations was founded in 1945 after the end of the Second World War, with the aim of promoting peace and cooperation.",
    #     "answers": {
    #         "text": ["1945"],
    #         "answer_start": [28]
    #     }
    # }
]

# Convert the data into a format suitable for the dataset
dataset = Dataset.from_dict({
    "question": [item["question"] for item in data_list],
    "context": [item["context"] for item in data_list],
    "answers": [item["answers"] for item in data_list]
})

def prepare_train_features(examples):
    tokenized_inputs = tokenizer(
        examples["question"],
        examples["context"],
        max_length=384,
        truncation="only_second",
        padding="max_length",
        return_offsets_mapping=True,
        return_tensors="pt"
    )

    # Initialize lists to store the computed token start and end positions
    start_positions = []
    end_positions = []

    for i in range(len(examples["question"])):
        # Extract the context, question, and answer from the current example
        context = examples["context"][i]
        answer = examples["answers"][i]
        start_char = answer["answer_start"][0]
        end_char = start_char + len(answer["text"][0])

        # Get the offset mapping for this example
        offset_mapping = tokenized_inputs["offset_mapping"][i]

        # Find the start and end token index for the answer
        sequence_ids = tokenized_inputs.sequence_ids(i)
        start_token_index = None
        end_token_index = None

        for j, (offset, seq_id) in enumerate(zip(offset_mapping, sequence_ids)):
            if seq_id == 1 and offset[0] <= start_char and offset[1] > start_char:
                start_token_index = j
            if seq_id == 1 and offset[0] < end_char and offset[1] >= end_char:
                end_token_index = j
                break

        # Handle cases where no valid answer is found within the span
        if start_token_index is None or end_token_index is None:
            start_token_index = tokenizer.model_max_length
            end_token_index = tokenizer.model_max_length

        start_positions.append(start_token_index)
        end_positions.append(end_token_index)

    tokenized_inputs["start_positions"] = start_positions
    tokenized_inputs["end_positions"] = end_positions

    # Remove the offset mapping to avoid issues during training
    del tokenized_inputs["offset_mapping"]

    return tokenized_inputs


# Tokenize and align the input data
tokenized_dataset = dataset.map(prepare_train_features, batched=True)

# Define training arguments
training_args = TrainingArguments(
    output_dir='./results',
    num_train_epochs=3,
    per_device_train_batch_size=4,
    warmup_steps=500,
    weight_decay=0.01,
    logging_dir='./logs'
)

# Initialize the trainer
trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset
)

# Train the model
trainer.train()

model.save_pretrained('./my_fine_tuned_model')
tokenizer.save_pretrained('./my_fine_tuned_model')
```
